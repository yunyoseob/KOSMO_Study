form 태그 속성

-----------------

name	자바스크립트, jQuery, CSS로 폼을 제어할 때 사용하는 폼 이름 지정

id		자바스크립트, jQuery, CSS자바스크립트로 폼을 제어할 때 사용하는 폼 이름 지정

method	폼 요소 내용을 서버쪽 프로그램에 넘겨주는 방식 지정 


		get	256 ~ 4096byte 서버로 전송 가능
			웹 브라우저 주소 표시줄에 사용자가 입력한 내용이 표시 됨 
			http://localhost:8088/firstWeb/html/formGetTest.html?id=%EB%B0%95%EA%B1%B4%EC%9B%90&pw=1234#
			
			클라이언트가 요청한 정보가 http 프로토콜 요청 정보가
			요청 Header 부분에 담아서 전송된다.

		post 입력한 내용의 길이에 제한을 받지 않음 
			웹 브라우저 주소 표시줄에 내용 표시 않됨
			http://localhost:8088/firstWeb/html/formPostTest.html#

			클라이언트가 요청한 정보가 http 프로토콜 요청 정보가
			요청 body 부분에 담아서 전송된다.

		HTTP 프로토콜 요청 Request, 응답 Response
		
		요청	GET Request			
			----------------------
			Header
			http://localhost:8088/firstWeb/html/formGetTest.html

			id=%EB%B0%95%EA%B1%B4%EC%9B%90
			&
			pw=1234# <-- 쿼리 스트링 Query String key=value&key=value
			----------------------
			
			----------------------
			Body

			----------------------

		요청 POST Request			
			----------------------
			Header
			http://localhost:8088/firstWeb/html/formPostTest.html#

			----------------------
			
			----------------------
			Body

			id=%EB%B0%95%EA%B1%B4%EC%9B%90
			&
			pw=1234#
			----------------------

		응답 Response
			
			----------------------
			Header
			----------------------
			
			----------------------
			Body
			----------------------

action	form 태그 안의 내용을 처리해 줄 서버 프로그램을 지정


> **form태그 예제 : fromGetTestp.html**


```html
<!DOCTYPE html>
<html>
<head>
<meta charset="UTF-8">
<title>FORM TEST</title>
</head>
<body>
	<h2>FORM TEST GET 방식</h2>
	<hr>
	<form method="GET" action="#">
	<!-- method="GET"은 HEADER 부분에 표시된다. -->
	<!-- 웹 브라우저 주소 표시줄에 사용자가 입력한 내용이 표시됨. -->
	<!-- 아이디에 velog 입력 -->
	<!-- 비밀번호에 1234 입력 -->
	<!-- http://localhost:8088/firstWeb/html/
	     fromGetTestp.html?id=velog&pw=1234# -->
	<!-- 개발자 도구에서 Payload에 보면 Query String Parameters라고 적혀있다. -->
	<!-- 헤더의 쿼리string으로 들어간다. -->
	<!-- id<key> : 입력값<value> -->
	<!-- GET은 256~4096byte 서버로 전송가능 -->
		아이디 : <input type="text" name="id"><br>
		비밀번호 : <input type="text"name="pw"><br>
		<input type="submit" value="GET방식"><br>
		
	<!-- 아이디에 벨로그 
	 	  비밀번호에 그로벨 입력 -->
	<!-- http://localhost:8088/firstWeb/html/
		 fromGetTestp.html?
		 id=%EB%B2%A8%EB%A1%9C%EA%B7%B8
		 &
		 pw=%EA%B7%B8%EB%A1%9C%EB%B2%A8# -->
	
	<!-- %~~%은 EBCDIC(확장 이진화 십진법 교환 부호)로
		 표기되어 나타난 것이다. -->
		 
	<!-- form method="GET"은 개발자 도구에서
		 General에 Requset URL에 입력한
		  아이디와 비밀번호가  적혀있다.  -->
	 	  
	</form>
	<!-- 가장 마지막에 열려있는 웹 브라우저에서 열린다. -->
</body>
</html>
```

> **form 태그 예제 : formPostTestp.html**

```html
<!DOCTYPE html>
<html>
<head>
<meta charset="UTF-8">
<title>FORM TEST</title>
</head>
<body>
<h2>FORM TEST POST방식</h2>
<hr>

<form method="POST" action="#">
	<!-- 입력한 내용의 길이에 제한을 받지 않음  -->
	<!-- 웹 브라우저 주소 표시줄에 내용 표시 되지 않음 -->
	<!-- 아이디에 velog 입력 -->
	<!-- 비밀번호에 1234 입력 -->
	<!-- http://localhost:8088
	 	 /firstWeb/html/fromPostTestp.html# -->

	<!-- 클라이언트가 요청한 정보가 http 프로토콜 요청 정보가
		 요청 body 부분에 담아서 전송된다. -->
	<!-- id<key> : 입력값 <value> -->
	<!-- 개발자 도구에서 Payload를 보면
	Form Data라고 작성되어 있다. -->

	아이디 : <input type="text" name="id"><br>
	비밀번호 : <input type="text" name="pw"><br>
	<!-- 개발자도구에 Payload에 view source
		 id=velog&pw=1234 -->
	<!-- 개발자도구에 Payload에 view parsed
		 id=velog
		 pw=1234 -->
	
	<input type="submit" value="POST방식"><br>
</form>
</body>
</html>
```
